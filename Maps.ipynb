```python
import pandas as pd
import scipy.stats as st
import numpy as np
import hvplot.pandas
import geoviews as gv
import warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
path = "CSV/zip_codes.csv"
```

```python
zipcode_df = pd.read_csv(path)
```

```python
pd.options.display.max_columns = 700
zipcode_df.head()
```

```python
newcol = []
for i in range(len(zipcode_df["Date"])):
    newcol.append(int(zipcode_df["Date"][i].split("/")[2]))
zipcode_df["Year"] = newcol
zipcode_df

data_df = zipcode_df.loc[zipcode_df["Year"].isin([2018, 2019, 2020, 2021, 2022])]
data_df.head()
```

# FIVE YEAR DATA

```python
#Create smaller dataframe for plotting
small_df = data_df.loc[:, ["Crash Record ID", "Make", "Model", "Severity", "Damages", "Primary Contributory Cause",
                              "Weather Condition", "Lighting Condition",
                              "Injury Type", "Latitude", "Longitude", "Location", "Zip Code", "Time", "Year",
                             "Posted Speed Limit"]]
```

```python
small_df["Injury Type"].value_counts()
```

```python
fatal_df = small_df.loc[small_df["Injury Type"].isin(["FATAL"])]
fatal_df
```

```python
fatal_df["Make"].value_counts().head()
```

```python
len(fatal_df)
```

```python
five_year_map = fatal_df.hvplot.points(
    "Longitude",
     "Latitude",
    geo = True,
    tiles = "EsriStreet",
    frame_width = 450,
    frame_height = 635,
    scale = 0.8,
    size = 80,
    color = "Primary Contributory Cause",
    hover_cols = ["Make","Model", "Damages", "Time",
                 "Weather Condition", "Lighting Condition", "Zip Code"],
    title = ("Fatal Vehicle Collisions by Primary Contributory Cause (2018-2022)"),
    legend = "right",
    xlabel = "Longitude",
    ylabel = "Latitude"
    
)
# Display the map
five_year_map
```

# Map of fatal collisions by Make[CHEVROLET]

```python
chevy_df["Make"].value_counts()
```

# Past Year Comparisons

```python
last_year_df = small_df.loc[small_df["Year"].isin([2022])]

reduced_df = last_year_df.loc[last_year_df["Injury Type"].isin(["FATAL"])]
reduced_df.head()
```

```python
num_fatal = (len(reduced_df))
num_fatal
```

```python
%%capture --no-display

# Configure the map plot
last_year_fatal_map = reduced_df.hvplot.points(
    "Longitude",
     "Latitude",
    geo = True,
    tiles = "EsriStreet",
    frame_width = 450,
    frame_height = 500,
    scale = 0.8,
    size = 250,
    color = "Primary Contributory Cause",
    hover_cols = ["Make","Model", "Damages", "Time",
                 "Weather Condition", "Lighting Condition", "Zip Code"],
    xlabel = "Longitude",
    ylabel = "Latitude",
    title = ("Fatal Vehicle Collisions by Primary Contributory Cause (2022)")
)
# Display the map
last_year_fatal_map
```

```python
# 2022 crashes with nonincapacitating injuries 
reduced_noninc_df = last_year_df.loc[last_year_df["Injury Type"].isin(["NONINCAPACITATING INJURY"])]
reduced_noninc_df
```

```python
%%capture --no-display

# Configure the map plot
last_year_nonfatal_map = reduced_noninc_df.hvplot.points(
    "Longitude",
     "Latitude",
    geo = True,
    tiles = "EsriStreet",
    frame_width = 450,
    frame_height = 850,
    scale = 0.8,
    size = 50,
    color = "Primary Contributory Cause",
    xlabel = "Longitude",
    ylabel = "Latitude",
    hover_cols = ["Make","Model", "Damages", "Time",
                 "Weather Condition", "Lighting Condition", "Zip Code"],
)
# Display the map
last_year_nonfatal_map
```
